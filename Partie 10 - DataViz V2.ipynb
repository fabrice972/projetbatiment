```python
# Ces 2 graphs ne nous donnent pas d'information évidente. Un même bien peut figurer plusieurs fois sur ce graphique 
# puisqu’il a pu être revendu plusieurs fois avec une valeur foncière qui a évolué. Donc interprétation très compliquée.
# On va plutôt créer une variable sur l'évolution des valeurs foncières d'un même lot entre 2 mutations

df_selec_2[["etaban_lot","valeurfonc","anneemut", "qty_of_mut"]]
```

```python
df_selec_2["qty_of_mut"].value_counts()
```

```python
# On commence par isoler les lots qui ont été revendus au moins 1 fois pendant la période

t=df_selec_2[df_selec_2["qty_of_mut"]>0][["etaban_lot","anneemut","valeurfonc"]].sort_values(by=["etaban_lot","anneemut"])
```

```python
t.head()
```

```python
# On créé une colonne qui va nous permettre de distinguer la 1ère ligne de mutation (first)
# pour calculer l'évolution des prix sur les autres lignes (not first)

etaban_lot=t["etaban_lot"].iloc[0]
t["range_revente_first"]="not first"

for index, value in enumerate(t["etaban_lot"]) :
    if etaban_lot!=value :
        t["range_revente_first"].iloc[index]="first"
    etaban_lot=value
    print(value)
```

```python
t["range_revente_first"].value_counts()
```

```python
t.head()
```

```python
# Renommer la valeur de "range_revente_first" en 1ère ligne du df pq elle a servi à la mise en place du compteur
# Elle était donc restée en "not first"

t["range_revente_first"].iloc[0]="first"
```

```python
# Calcul du pourcentage d'évolution de la valeur foncière, d'une ligne à l'autre

t["Evol_val"]=t["valeurfonc"].pct_change()
```

```python
t.head()
```

```python
# Mise à 0 de la variable "Evol_val" pour le rang "first"

t["Evol_val"][t["range_revente_first"]=="first"]=0
```

```python
t.head()
```

```python
# Réduction du dataframe t aux colonnes "range_revente_first" et "Evol_val"

t=t[["range_revente_first","Evol_val"]]
```

```python
t
```

```python
# Fusion des dataframes df_selec_2 et t par un merge left sur l'index "Idmutation"

df_selec_2=df_selec_2.merge(right=t, on="idmutation", how="left")
```

```python
df_selec_2[["etaban_lot","anneemut","valeurfonc","range_revente_first", "Evol_val"]].sort_values(by=["etaban_lot","anneemut"])
```

```python
# Pour les biens qui n'ont pas été revendus pendant la période, nous allons établir une valeur estimée à partir
# du prix au m² standardisés des appartements à Paris au 4e trimestre 2021
# https://paris.notaires.fr/fr/carte-des-prix
df_selec_2.shape
```

```python
# Extraction en CSV de la page 10 de ce document : https://notairesdugrandparis.fr/sites/default/files/Prix%20T4%202021_0.pdf
prix_notaire=pd.read_csv("Carte_prix_notaires_idf.csv",sep=";", encoding= 'unicode_escape')
prix_notaire
```

```python
# Fusion des dataframes df_selec_2 et du prix_notaire par un merge left sur l'index "idmutation"

df_selec_2=df_selec_2.reset_index().merge(prix_notaire, on="l_codinsee", how="left").set_index('idmutation')
```

```python
# Création d'une variable "Valeur_estimée" :
df_selec_2["Valeur_estimée"]=df_selec_2["scarrez"]*df_selec_2["Estimation_prix_m2"]
```

```python
df_selec_2["Evol_val"].isna().sum()
```

```python
# La variable "Evol_val" = "nan" pour tous les biens qui n'ont pas été revendus => Remplacement de ces nan par le pourcentage
# d'évolution entre la valeur foncière initiale et la valeur estimée au 31/12/2021 selon carte notaires

df_selec_2["Evol_val"]=df_selec_2["Evol_val"].fillna((df_selec_2["Valeur_estimée"]-df_selec_2["valeurfonc"])/df_selec_2["valeurfonc"])
```

```python
# On obtient donc une variable Evol_val qui sera égale à :
# (Valeur estimée-Valeur foncière)/valeurfoncière pour tous les lots qui n'ont pas été revendus
# - 0 pour la 1ère mutation des lots qui ont été revendus
# - % d'augmentation pour les mutations suivantes des lots qui ont été revendus
```
