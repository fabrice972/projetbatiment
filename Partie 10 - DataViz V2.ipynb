```python
# Ces 2 graphs ne nous donnent pas d'information Ã©vidente. Un mÃªme bien peut figurer plusieurs fois sur ce graphique 
# puisquâ€™il a pu Ãªtre revendu plusieurs fois avec une valeur fonciÃ¨re qui a Ã©voluÃ©. Donc interprÃ©tation trÃ¨s compliquÃ©e.
# On va plutÃ´t crÃ©er une variable sur l'Ã©volution des valeurs fonciÃ¨res d'un mÃªme lot entre 2 mutations

df_selec_2[["etaban_lot","valeurfonc","anneemut", "qty_of_mut"]]
```

```python
df_selec_2["qty_of_mut"].value_counts()
```

```python
# On commence par isoler les lots qui ont Ã©tÃ© revendus au moins 1 fois pendant la pÃ©riode

t=df_selec_2[df_selec_2["qty_of_mut"]>0][["etaban_lot","anneemut","valeurfonc"]].sort_values(by=["etaban_lot","anneemut"])
```

```python
t.head()
```

```python
# On crÃ©Ã© une colonne qui va nous permettre de distinguer la 1Ã¨re ligne de mutation (first)
# pour calculer l'Ã©volution des prix sur les autres lignes (not first)

etaban_lot=t["etaban_lot"].iloc[0]
t["range_revente_first"]="not first"

for index, value in enumerate(t["etaban_lot"]) :
    if etaban_lot!=value :
        t["range_revente_first"].iloc[index]="first"
    etaban_lot=value
    print(value)
```

```python
t["range_revente_first"].value_counts()
```

```python
t.head()
```

```python
# Renommer la valeur de "range_revente_first" en 1Ã¨re ligne du df pq elle a servi Ã  la mise en place du compteur
# Elle Ã©tait donc restÃ©e en "not first"

t["range_revente_first"].iloc[0]="first"
```

```python
# Calcul du pourcentage d'Ã©volution de la valeur fonciÃ¨re, d'une ligne Ã  l'autre

t["Evol_val"]=t["valeurfonc"].pct_change()
```

```python
t.head()
```

```python
# Mise Ã  0 de la variable "Evol_val" pour le rang "first"

t["Evol_val"][t["range_revente_first"]=="first"]=0
```

```python
t.head()
```

```python
# RÃ©duction du dataframe t aux colonnes "range_revente_first" et "Evol_val"

t=t[["range_revente_first","Evol_val"]]
```

```python
t
```

```python
# Fusion des dataframes df_selec_2 et t par un merge left sur l'index "Idmutation"

df_selec_2=df_selec_2.merge(right=t, on="idmutation", how="left")
```

```python
df_selec_2[["etaban_lot","anneemut","valeurfonc","range_revente_first", "Evol_val"]].sort_values(by=["etaban_lot","anneemut"])
```

```python
# Pour les biens qui n'ont pas Ã©tÃ© revendus pendant la pÃ©riode, nous allons Ã©tablir une valeur estimÃ©e Ã  partir
# du prix au mÂ² standardisÃ©s des appartements Ã  Paris au 4e trimestre 2021
# https://paris.notaires.fr/fr/carte-des-prix
df_selec_2.shape
```

```python
# Fusion des dataframes df_selec_2 et du prix_notaire par un merge left sur l'index "idmutation"

df_selec_2=df_selec_2.reset_index().merge(prix_notaire, on="l_codinsee", how="left").set_index('idmutation')
```

```python
# CrÃ©ation d'une variable "Valeur_estimÃ©e" :
df_selec_2["Valeur_estimÃ©e"]=df_selec_2["scarrez"]*df_selec_2["Estimation_prix_m2"]
```

```python
df_selec_2["Evol_val"].isna().sum()
```

```python
# La variable "Evol_val" = "nan" pour tous les biens qui n'ont pas Ã©tÃ© revendus => Remplacement de ces nan par le pourcentage
# d'Ã©volution entre la valeur fonciÃ¨re initiale et la valeur estimÃ©e au 31/12/2021 selon carte notaires

df_selec_2["Evol_val"]=df_selec_2["Evol_val"].fillna((df_selec_2["Valeur_estimÃ©e"]-df_selec_2["valeurfonc"])/df_selec_2["valeurfonc"])
```

```python
# On obtient donc une variable Evol_val qui sera Ã©gale Ã  :
# (Valeur estimÃ©e-Valeur fonciÃ¨re)/valeurfonciÃ¨re pour tous les lots qui n'ont pas Ã©tÃ© revendus
# - 0 pour la 1Ã¨re mutation des lots qui ont Ã©tÃ© revendus
# - % d'augmentation pour les mutations suivantes des lots qui ont Ã©tÃ© revendus
```
