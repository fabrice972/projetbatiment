```python
# A 98%, une mutation concerne un seul lot. 
# Moins de 2% des mutations couvrent plusieurs lots  
# On va donc garder une ligne par mutation (la premiÃ¨re)  
```

```python
dico = {}
for col in selec_scarrez_sans_doublon.columns:
    dico[col]='first'
```

```python
selec_scarrez_sans_doublon = selec_scarrez_sans_doublon.groupby("idmutation", as_index=False).agg(dico)
```

```python
selec_scarrez_sans_doublon.shape
```

```python
# On renomme notre dataframe pour simplification
df=selec_scarrez_sans_doublon
```

```python
df[["nolot","scarrez", "datemut", "idmutation"]]
```

```python
# On se rend compte que le numÃ©ro de lot n'identifie pas le lot. Il faut y ajouter l'Etaban
```

```python
# CrÃ©ation d'une colonne avec etaban + numÃ©ro de lot
```

```python
df['etaban_lot']=df['etaban']+"_"+df['nolot'].astype('str')
```

```python
df['etaban_lot']
```

```python
# CrÃ©ation de la variable cible 
# Recherche du nombre de mutations par lot
```

```python
df1=df['etaban_lot'].value_counts().to_frame()
```

```python
df1=df1.reset_index()
```

```python
df1=df1.rename(columns={'index':'etaban_lot','etaban_lot':'qty_of_mut' })
```

```python
df1.head()
```

```python
df1['qty_of_mut'].value_counts()
```

```python
df1['y']= df1['qty_of_mut'].apply(lambda x : 1 if x>1 else 0)
```
